Arctic ice maps
===============

In this problem set some pictures of the artic ice are given and should be analyzed and presented.

Starting the problem set with the initial imports of the used libraries...

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import glob
```

Loading now the first dataset/image dated 15th April 2008.   
Using the White/Blue-scale to show the concetration of ice in the pixels...

```python
data = np.load('ice_data/20080415.npy')

plt.figure(figsize=(10,10))

cpicture = plt.imshow(data, origin='lower', cmap=plt.cm.Blues_r)

cbar = plt.colorbar(cpicture)
plt.axis('off')
```

Seeing the ice (nearly white), the ocean (blue) and the landscape (in data NaN and display transparent, unfortunately the background color is also white...).

Now I am scanning the directory 'ice_data' and sorting the found files.

```python
files = glob.glob('ice_data/*.npy')
files.sort()
```

Starting the analyze of the datasets with initializing an numpy array filled with zeros.  
Loading the data of each set, extracting the date from the filename, calculate a year based time variable and count the pixels with ice concentration above 50%.

```python
pixels_above_50 = np.zeros([len(files), 2])
i = 0

for file in files:
    data = np.load(file)

    year = int(file[9:13])
    month = int(file[13:15])
    day = int(file[15:17])

    time = year + (month - 1 + (day - 1) / 30.0) / 12.0

    pixels_above_50[i,0] = time
    pixels_above_50[i,1] = np.size(data[data >= 50])

    i = i + 1
```

Setting an other formatter for the x-axis-scale makes it more beautiful.  
In plot you see on x-axis time and y-axis the pixel count.

It shows you that there are summers and winters periodically ;)

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111)
ax.set_xlabel('time in years')
ax.set_ylabel('pixel count')
ax.set_title('Progress of the count of pixels with more than 50% of ice')
fig.gca().xaxis.set_major_formatter(ScalarFormatter(useOffset=False))
ax.plot(pixels_above_50[:,0], pixels_above_50[:,1])
```

There is problem with this analyze:  
The pixels in our datasets do not represent the same area due to the spherical surface.

This map shows the size of an pixel:

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
area = np.load('ice_data_area.npy')
cax = ax.imshow(area, origin='lower')
fig.colorbar(cax)
ax.axis('off')
```

Now we are going to calculate effective coverage of the ice.  

Extracting the date of image, multiply the two images, calculate the total sum of ice and only pixel with more than 99% ice coverage.

```python
covered_area = np.zeros([len(files), 3])
i = 0

for file in files:
    data = np.load(file)

    year = int(file[9:13])
    month = int(file[13:15])
    day = int(file[15:17])

    time = year + (month - 1 + (day - 1) / 30.0) / 12.0

    covered_area[i, 0] = time
    tmp = data / 100.0 * area
    covered_area[i, 1] = np.nansum(tmp)
    covered_area[i, 2] = np.sum(tmp[data > 99])

    i = i + 1
```

Getting a little bit different curves than just in the plot above:

```python
fig = plt.figure(figsize=(8,5))
ax = fig.add_subplot(111)

ax.set_xlabel('time in years')
ax.set_ylabel('ice coverage in km^2')

fig.gca().xaxis.set_major_formatter(ScalarFormatter(useOffset=False))
ax.plot(covered_area[:, 0], covered_area[:, 1])
ax.plot(covered_area[:, 0], covered_area[:, 2])
```

The blue curve shows the total ice coverage and the green one only the ice coverage of pixels with more than 99% ice in.

We are looking now for the date, where the ice coverage of pixels with more than 99%, has a minimum.  
Reveal the date an loading a dataset of the month and day, but from 2006.

```python
time = float(covered_area[covered_area[:,2]==np.min(covered_area[:,2]), 0])
area = float(covered_area[covered_area[:,0]==time, 2])

year = int(time)
month = int((time - int(time)) * 12 + 1)
tmp = (time - int(time)) * 12 + 1
day = int((tmp - int(tmp)) * 30 + 1)

data2006 = np.load("ice_data/{year:04d}{month:02d}{day:02d}.npy".format(year=2006, month=month, day=day))
dataYear = np.load("ice_data/{year:04d}{month:02d}{day:02d}.npy".format(year=year, month=month, day=day))

"{year:04d}/{month:02d}/{day:02d} the ice coverage was ~{area:.1f}km^2".format(year=year, month=month, day=day, area=area)
```

```python
fig = plt.figure(figsize=(16,8))
ax1 = fig.add_subplot(121)
cax1 = ax1.imshow(data2006, origin='lower', cmap=plt.cm.Blues_r)
ax2 = fig.add_subplot(122)
cax2 = ax2.imshow(dataYear, origin='lower', cmap=plt.cm.Blues_r)
ax1.set_title('2006')
ax2.set_title(str(year))
ax1.axis('off')
ax2.axis('off')
```

The two images show the loss of ice between 2006 and 2011.

Better is a difference map:

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
cax = ax.imshow(dataYear - data2006, origin='lower', cmap=plt.cm.RdBu)
fig.colorbar(cax)
ax.axis('off')
```

Now calculate the average off 2006 and 2011:

```python
files2006 = glob.glob('ice_data/2006*.npy')
files2011 = glob.glob('ice_data/2011*.npy')
files2006.sort()
files2011.sort()

data2006 = range(len(files2006))
for i in data2006:
    data2006[i] = np.load(files2006[i])

data2011 = range(len(files2011))
for i in data2011:
    data2011[i] = np.load(files2011[i])

data2006 = np.array(data2006)
data2011 = np.array(data2011)

data2006 = np.ma.masked_array(data2006,np.isnan(data2006))
data2011 = np.ma.masked_array(data2011,np.isnan(data2011))

mean2006 = np.mean(data2006, axis=0)
mean2011 = np.mean(data2011, axis=0)

mean2006 = mean2006.filled(np.nan)
mean2011 = mean2011.filled(np.nan)

fig = plt.figure(figsize=(16,8))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
cax1 = ax1.imshow(mean2006, origin='lower', cmap=plt.cm.Blues_r)
cax2 = ax2.imshow(mean2011, origin='lower', cmap=plt.cm.Blues_r)
ax1.set_title('2006')
ax2.set_title('2011')
ax1.axis('off')
ax2.axis('off')
```
